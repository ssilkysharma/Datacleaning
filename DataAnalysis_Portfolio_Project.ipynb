```python

from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
import pandas as pd
df = pd.read_csv('FF_Race_50.csv')
```

```python
import os
print(os.listdir('.'))
```

```python
df.head()
```

```python
df['Full name'] = df['First'] + ' ' + df['Last']
df.head()
```

```python
df['Time'] = pd.to_timedelta(df['Time'])
```

```python
df['Total_Minutes'] = df['Time'].dt.total_seconds() / 60
```

```python
df['Total_Minutes'] = df['Total_Minutes'].round().astype(int)
```

```python
df.head()
```

```python
df.rename(columns={'Division': 'Gender'}, inplace=True)
```

```python
def get_lat_long(city,state):
  address = f"{city}, {state}"
  try:
    geolocator = Nominatim(user_agent="running", timeout=10)
    location = geolocator.geocode(address)
    if location:
      return location.latitude,location.longitude
    else:
      return None,None
  except GeocoderTimedOut:
    return get_lat_long(city,state)
```

```python
df['latitude'],df['longitude'] = zip(*df.apply(lambda row: get_lat_long(row['City'],row['State']),axis=1))
```

```python
df.head()
```

```python
df['latlong'] = df['latitude'].astype(str) + ','+ df['longitude'].astype(str)
```

```python
df.to_csv('ultracleanedupdata.csv',index=False)
```
